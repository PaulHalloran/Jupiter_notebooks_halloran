```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import matplotlib.cm as cm
import matplotlib.pyplot as mpl

mpl.rcParams['font.size'] = 14

hfont = {'fontname':'Arial'}
```

```python
# file = '/Users/ph290/Downloads/diameter_circ_data.txt'
file = '/Users/ph290/Downloads/dia_circ.txt'
data = pd.read_csv(file,header=0,delimiter=' ')

file2 = '/Users/ph290/Downloads/Figure4.xlsx'
data2 = pd.read_excel(file2,sheet_name='All',header=1)

data3 = pd.read_excel(file2,sheet_name='R1',header=1)

data2b = data2.loc[~(data2['Diameter_Object(M01, Ch01, Tight)']==0.0)]
data3b = data3.loc[~(data3['Diameter_Object(M01, Ch01, Tight)']==0.0)]
```

```python
x,y=data2b['Diameter_Object(M01, Ch01, Tight)'],data2b['Circularity_Object(M01, Ch01, Tight)']
# Calculate the point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
```

```python

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(x, y, c=z, s=10, edgecolor='', cmap=cm.plasma)
plt.ylim([0,30])
plt.xlim([0,40])
plt.xlabel('Diameter ($\mu$m)',**hfont)
plt.ylabel('Circularity',**hfont)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.savefig('/Users/ph290/Downloads/for_beth2.png',dpi=600)
plt.show()
```
