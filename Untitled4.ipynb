```python
MSEs = []
for len in np.arange(1,1000):
    x = np.arange(len)
    y1 = np.random.rand(len)
    y2 = np.zeros(len)+0.5
#     plt.plot(x,y1)
#     plt.plot(x,y2)
    MSE = np.nanmean((y2-y1)**2) 
    MSEs = np.append(MSEs,MSE)
```
