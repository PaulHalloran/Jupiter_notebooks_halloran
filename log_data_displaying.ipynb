```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from sklearn.datasets import make_regression
```

### Load data

```python
x = np.load('/Users/ph290/Downloads/n1_Bell15_binsizes.npy')
y = np.load('/Users/ph290/Downloads/n1_Bell15_mse.npy')
```

### Plot raw data

```python
plt.scatter(x,y)
```

### Plot on log log axes

```python
plt.loglog(x,y)
```

### Log data then plot

```python
x2 = np.log(x)
y2 = np.log(y)

plt.scatter(x2,y2)
```

### log data then plot, but make x and y axes relate to the raw data
### i.e. equivilent to 'Plot on log log axes'

```python
x2 = np.log(x)
y2 = np.log(y)

fig, ax = plt.subplots()
ax.scatter(x2,y2)

xmin, xmax = ax.get_xlim()
xticks = np.linspace(xmin,xmax,10)
ax.set_xticks(xticks)
ax.set_xticklabels(np.exp(xticks).round(2), rotation=45)

ymin, ymax = ax.get_ylim()
yticks = np.linspace(ymin,ymax,10)
ax.set_yticks(yticks)
ax.set_yticklabels(np.exp(yticks).round(2), rotation=45)
```

### As above but neater

```python
x2 = np.log(x)
y2 = np.log(y)

fig, ax = plt.subplots()
ax.scatter(x2,y2)

xmin, xmax = ax.get_xlim()
xticks = np.linspace(xmin,xmax,10)
ax.set_xticks([np.log(np.power(10,1)),np.log(np.power(10,2)),np.log(np.power(10,3))])
ax.set_xticklabels([np.power(10,1),np.power(10,2),np.power(10,3)], rotation=45)

ymin, ymax = ax.get_ylim()
yticks = np.linspace(ymin,ymax,10)
ax.set_yticks(yticks)
ax.set_yticklabels(np.exp(yticks).round(2), rotation=45)
```
