```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.load('/Users/ph290/Downloads/250_Herr07_binsizes.npy')
y = np.load('/Users/ph290/Downloads/250_Herr07_mse.npy')

# I'm using a -1.5 power base 

# x_pwr = np.power(x,-1.5)
# y_pwr = np.power(y,-1.5)
```

```python

fig, ax = plt.subplots()

ax.scatter(x,y,s=3)
ax.set_xlim([0,12])
ax.set_ylim([0.75,2.5])
```

```python
power = -1.5

fig, ax = plt.subplots()

x2 = np.power(x,power)
y2 = np.power(y,power)
ax.scatter(x2,y2,s=3)
```

```python
power = -1.5

fig, ax = plt.subplots()

x2 = np.power(x,power)
y2 = np.power(y,power)
ax.scatter(x2,y2,s=3)

xmin, xmax = ax.get_xlim()
xticks = np.linspace(xmin,xmax,10)
ax.set_xticks(xticks)
ax.set_xticklabels(np.power(xticks,(1.0/power)).round(2), rotation=45)

ymin, ymax = ax.get_ylim()
yticks = np.linspace(ymin,ymax,10)
ax.set_yticks(yticks)
ax.set_yticklabels(np.power(yticks,(1.0/power)).round(2), rotation=45)
```

```python
power = -1.5

fig, ax = plt.subplots()

x2 = np.power(x,power)
y2 = np.power(y,power)
ax.scatter(np.flip(x2),np.flip(y2),s=3)

xmin, xmax = ax.get_xlim()
ax.set_xlim([xmax,xmin])
xticks = np.linspace(xmin,xmax,10)
ax.set_xticks(xticks)
ax.set_xticklabels(np.power(xticks,(1.0/power)).round(2), rotation=45)

ymin, ymax = ax.get_ylim()
ax.set_ylim([ymax,ymin])
yticks = np.linspace(ymin,ymax,10)
ax.set_yticks(yticks)
ax.set_yticklabels(np.power(yticks,(1.0/power)).round(2), rotation=45)
```
