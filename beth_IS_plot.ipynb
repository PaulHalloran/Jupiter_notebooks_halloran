```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import matplotlib.cm as cm
import matplotlib.pyplot as mpl

mpl.rcParams['font.size'] = 14

hfont = {'fontname':'Arial'}
```

```python
# file = '/Users/ph290/Downloads/diameter_circ_data.txt'
file = '/Users/ph290/Downloads/dia_circ.txt'
data = pd.read_csv(file,header=0,delimiter=' ')
```

```python
plt.scatter(data.Diameter_M01,data.Circularity_M01)
```

```python
x,y=data.Diameter_M01.values,data.Circularity_M01.values
# Calculate the point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)
```

```python


fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(x, y, c=z, s=10, edgecolor='', cmap=cm.plasma)
plt.ylim([0,30])
plt.xlim([0,40])
plt.xlabel('Diameter ($\mu$m)',**hfont)
plt.ylabel('Circularity',**hfont)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.savefig('/Users/ph290/Downloads/for_beth.png',dpi=600)
plt.show()
```

```python
file2a = '/Users/ph290/Downloads/sediment.txt'
file2b = '/Users/ph290/Downloads/sediment_liths.txt'
file2c = '/Users/ph290/Downloads/sediment_spheres.txt'
data_sed = pd.read_csv(file2a,header=0,delimiter='\t')
data_lith = pd.read_csv(file2b,header=0,delimiter='\t')
data_sphere = pd.read_csv(file2c,header=0,delimiter='\t')
```

```python
xa,ya=data_sed.Diameter_M01.values,data_sed.Circularity_M01.values
# Calculate the point density
xy = np.vstack([xa,ya])
za = gaussian_kde(xy)(xy)

xb,yb=data_sphere.Diameter_M01.values,data_sphere.Circularity_M01.values
# Calculate the point density
xy = np.vstack([xb,yb])
zb = gaussian_kde(xy)(xy)

xc,yc=data_lith.Diameter_M01.values,data_lith.Circularity_M01.values
# Calculate the point density
xy = np.vstack([xc,yc])
zc = gaussian_kde(xy)(xy)
```
