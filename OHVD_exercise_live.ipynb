## An example from the ocean

## Introduction

The circulation of the ocean is 'driven' by density contrasts. Dense water sinks below less-dense water, and vice versa.

The Mediterranean is a very warm (warm = low density) but salty (salty = high density) ocean basin. Water 'spills' out of the Mediterranean into the open Atlantic Ocean, carrying a lot of heat withh it. I want you to try and identify at what depths sea-water originating from the Mediterranean can be found in the Atlantic Ocean.
